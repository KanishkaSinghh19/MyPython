```python
choice = int(input())
match choice:
    case 1:
        print('one')
    case 2:
        print('two')
    case 3:
        print('three')
    case _:
        print('invalid')
```

```python
aa=int(input('num1'))
bb=int(input('num2'))
op=input('operation')
match op:
    case '+':
        print(aa+bb)
    case '-':
        print(aa-bb)
    case '*':
        print(aa*bb)
    case _:
        print('invalid op')
```

```python
day = input('enter day')
match day:
    case 'sat' | 'sun':
        print('weekend')
    case 'mon' | 'tue' | 'wed' | 'thu' | 'fri':
        print('weekday')
    case _:
        print('invalid entry')
```
